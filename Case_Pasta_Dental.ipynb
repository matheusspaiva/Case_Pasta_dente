# Importação

```python
import requests as r
from PIL import Image 
from IPython.display import display 
```

# DADOS

```python
#dados reis 
populacao = 213000000 #população do brasil 
escovacao_total = 53 #escovação completa 
escovacao_parcial = 31 #escova entre uma ou duas vezes ao dia
resto = 100 - (escovacao_parcial+escovacao_total)
#dados supostos
creme_dental = 90 #quantidade que vem em cada pasta 
```

```python
v3 =(escovacao_total*populacao/100)
v1 =(escovacao_parcial*populacao/100)
v = (resto*populacao/100)
```

```python
lista_dia = []
lista_consumo =[]
dia = 0
consumo = 0
while consumo != creme_dental:
  dia= dia+1
  consumo= consumo+3
  lista_consumo.append(consumo)
  lista_dia.append(dia)
```

```python
print(lista_consumo)
print(lista_dia)
```

```python
creme_dental = 90
lista_dia1 = []
lista_consumo1 =[]
dia1 = 0
consumo1 = 0
while consumo1 != creme_dental:
  dia1= dia1 + 1
  consumo1= consumo1 + 1
  lista_consumo1.append(consumo1)
  lista_dia1.append(dia1)
  
```

```python
print(lista_consumo1)
print(lista_dia1)
```
